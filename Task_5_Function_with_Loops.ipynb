## Problem Statement 1

Create a function to analyze engagement on a social media post.

### Rules
- Calculate total likes
- If Total Likes ≥ 1000 → Viral Post
- Otherwise → Normal Engagement

```python
def analyze_engagement(likes_list):
  total_likes = 0
  for likes in likes_list:
    total_likes += likes

  if total_likes >= 1000:
    status = "Viral Post"
  else:
    status = "Normal Engagement"

  return total_likes, status

likes_data = [200, 300, 150, 400]
total, post_status = analyze_engagement(likes_data)
print("Total Likes: ",total)
print("Post Status: ",post_status)
```

## Problem Statement 3

Create a function that analyzes seasonal rainfall data.

### Rules
- Calculate average rainfall
- If Average ≥ required level → Adequate Rainfall
- Otherwise → Inadequate Rainfall

```python
def check_rainfall(rainfall_list, required_level):
  total = 0
  for value in rainfall_list:
    total += value
  # Average rainfall
  average = total/len(rainfall_list)

  if average >= required_level:
    status = "Adequate Rainfall"
  else:
    status = "Inadequate Rainfall"
  return average, status

rainfall_data = [70, 75, 68, 80, 67]  # example
required = 75
avg_rainfall, rainfall_status = check_rainfall(rainfall_data, required)
print(f"Average rainfall: {avg_rainfall}")
print(f"Rainfall Status: {rainfall_status}")
```

## Problem Statement 4

Create a function to detect duplicate usernames in a system.

### Rules
- If duplicates exist → Flag accounts
- Else → Accounts are unique


```python
def detect_duplicates(usernames):
  unique_users = set(usernames)

  if len(unique_users) != len(usernames):
    return "Duplicate Accounts Found: Yes"
  else:
    return "Duplicate Accounts Found: No"

user_list = ["ravi123", "anita45", "ravi123", "kiran99"]
result = detect_duplicates(user_list)
print(result)
```

## Problem Statement 6

Create a function that filters crop prices above a premium threshold.

### Rules
- Price > ₹2000 → Premium Crop
- Else → Regular Crop

```python
def filter_premium_crops(price_list):
  premium = []

  for price in price_list:
    if price > 2000:
      premium.append(price)
  return premium

crop_prices = [1500, 2500, 1800, 3200, 1900]
premium_list = filter_premium_crops(crop_prices)
print(f"Premium Crops: {premium_list}")
```

## Problem Statement 7

Create a function to evaluate application health based on error count.

### Rules
- Errors = 0 → Healthy
- Errors ≤ 5 → Minor Issues
- Errors > 5 → Critical Issues

## Problem Statement 9

Create a function to calculate attendance percentage and check exam eligibility.

### Rules
- Attendance ≥ 75% → Eligible
- Attendance < 75% → Not Eligible

```python
def check_eligibility(attendance_list):
  present_count = 0
  for status in attendance_list:
    if status == "P":
      present_count += 1

  percentage = (present_count/len(attendance_list)) * 100

  if percentage >= 75:
    eligibility = "Eligible"
  else:
    eligibility = "Not Eligible"
  return percentage, eligibility

attendance_data = ["P", "P", "A", "P", "P"]
percent, status = check_eligibility(attendance_data)
print("Attendance Percentage:", percent)
print("Exam Eligibility:", status)
```

## Problem Statement 10

Create a function that calculates the electricity bill and classifies usage level.

### Billing Rules
- First 100 units → ₹3 per unit  
- Next 100 units (101–200) → ₹5 per unit  
- Above 200 units → ₹7 per unit  

### Usage Classification
- Total bill < ₹500 → Low Usage  
- ₹500 – ₹1500 → Moderate Usage  
- Total bill > ₹1500 → High Usage

```python
def calculate_electricity_bill(units):
    bill = 0

    if units <= 100:
      bill = units * 3
    elif units <= 200:
      bill = (100 * 3) + ((units - 100) * 5)
    else:
      bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
      status = "Low Usage"
    elif bill <= 1500:
      status = "Moderate Usage"
    else:
      status = "High Usage"

    return bill, status

units_consumed = 250
total_bill, usage_status = calculate_electricity_bill(units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Status:", usage_status)
```
