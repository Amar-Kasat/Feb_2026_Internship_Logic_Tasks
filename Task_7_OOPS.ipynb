## Problem Statement 1

Create a class to control a smart light that can be turned ON or OFF and display its status.

```python
class SmartLight:
  def __init__(self,name):
    self.name = name
    self.status = "OFF"
  # Turn ON LED
  def turn_on(self):
    self.status = "ON"
  # Turn OFF LED
  def turn_off(self):
    self.status = "OFF"
  # Display status
  def show_status(self):
    print(f"{self.name} is {self.status}")

# Object Creation
light = SmartLight("Bedroom Light")
light.turn_on()
light.show_status()
```

## Problem Statement 2

Create a class that stores employee name, ID, and department
and displays ID card details.

```python
class Employee:
  def __init__(self, name, id, dept):
    self.name = name
    self.id = id
    self.dept = dept
  # Display details
  def display(self):
    print("Employee ID Card")
    print(f"Name: {self.name}")
    print(f"ID: {self.id}")
    print(f"Department: {self.dept}")

# Object creation
emp = Employee("Rahul", "EMP102", "IT")
emp.display()
```

## Problem Statement 3

Create a class to store contact name and phone number
and display contact information.

```python
class Record:
  def __init__(self, name, number):
    self.name = name
    self.number = number
  # Display information
  def show_info(self):
    print("Contact Saved")
    print(f"Name: {self.name}")
    print(f"Phone: {self.number}")

# Object creation
rec = Record("Anita", 9876543210)
rec.show_info()
```

## Problem Statement 4

Create a class that stores product name and price
and prints a formatted price tag.

```python
class Product:
  def __init__(self, name, price):
    self.name = name
    self.price = price
  # Display price tag
  def show_tag(self):
    print(f"Product: {self.name}")
    print(f"Price: ₹{self.price}")

# Object creation
price = Product("Headphones", 2499)
price.show_tag()
```

## Problem Statement 5

Create a class that stores movie name and rating
and displays movie details.

```python
class Rating:
  def __init__(self, name, rating):
    self.name = name
    self.rating = rating
  # Display movie details
  def show_details(self):
    print(f"Movie: {self.name}")
    print(f"Rating: {self.rating}/5")

# Object creation
movie = Rating("Inception", 4.8)
movie.show_details()
```

## Problem Statement 6

Create a class that stores customer name and delivery address
and prints delivery details.

```python
class Customer:
  def __init__(self, name, address):
    self.name = name
    self.address = address
  # Delivery Details
  def details(self):
    print("Delivery Details")
    print(f"Customer: {self.name}")
    print(f"Address: {self.address}")

# Object creation
delivery = Customer("Suman", "Hyderabad")
delivery.details()
```
