## Problem Statement 1

A company evaluates employee performance scores at the end of the year.
We are given a dictionary containing employee names and their performance scores.

### Requirements
- Identify the highest performance score.
- Handle ties if multiple employees have the same highest score.
- Display all employees eligible for the top performance bonus.


```python
# Employee performance data
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# Find highest performance score
highest_score = max(employees.values())
# Find all employees with highest score
top_performers = []

for name, score in employees.items():
  if score == highest_score:
    top_performers.append(name)

  print("Top Performers Eligible for Bonus:", ", ".join(top_performers),f"(Score: {highest_score})")
```

## Problem Statement 2

An e-commerce website stores customer search queries.
Given a search query sentence, perform keyword analysis.

### Requirements
- Convert the input to lowercase.
- Ignore common punctuation.
- Count the frequency of each keyword.
- Display only keywords searched more than once.


```python
import string
query = "Buy mobile phone buy phone online"
# Convert to lowercase
query = query.lower()
# Remove punctuation
for ch in string.punctuation:
  query = query.replace(ch,"")
# Split into words
words = query.split()
# Count frequency
frequency = {}
for word in words:
  if word in frequency:
    frequency[word] += 1
  else:
    frequency[word] = 1
# Keep only words appearing more than once
result = {}
for word, count in frequency.items():
  if count > 1:
    result[word] = count
print(result)
```

## Problem Statement 3

A factory collects sensor readings every hour. Each reading is stored in a list where:
- Index → represents the hour
- Value → represents the sensor reading

### Requirements
- Identify readings that are even numbers (valid readings).
- Store them as (hour_index, reading_value) pairs.
- Ignore odd readings (invalid readings).


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
# Check each reading with its index
for hour, reading in enumerate(sensor_readings): # enumerate coverts reading into (idx,val) pairs
   if reading % 2 == 0:
    valid_readings.append((hour, reading))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## Problem Statement 4

A company wants to analyze which email providers its users are using.

### Requirements
- Count how many users belong to each email domain.
- Calculate the percentage usage of each domain.


```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}
# count domains
for email in emails:
  parts = email.split("@")
  domain = parts[1]

  if domain in domain_count:
    domain_count[domain] += 1
  else:
    domain_count[domain] = 1

# Calculate percentage
total = len(emails)
for domain in domain_count:
  percent = (domain_count[domain]/total) * 100
  print(domain,":",int(percent),"%")
```

## Problem Statement 5

A retail company tracks daily sales. Sudden spikes in sales may indicate promotions or unusual activity.

### Requirements
- Calculate the average daily sales.
- Detect days where sales are more than 30% above average.
- Display the day number and sale value.


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
# calculate average
average_sales = sum(sales) / len(sales)
# calculate spike threshold 30% above average
threshold = average_sales * 1.2
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

## Problem Statement 6

A system stores user IDs during registration. Duplicate IDs can cause data integrity issues.

### Requirements
- Identify duplicate user IDs.
- Display how many times each duplicate appears.


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
# Count frequency of each user id
frequency = {}
for user in user_ids:
  if user in frequency:
    frequency[user] += 1
  else:
    frequency[user] = 1

for user, count in frequency.items():
  if count > 1:
    print(f"{user} → {count} times")
```
